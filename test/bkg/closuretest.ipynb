# Perform closure test for ttbar All-Hadronic Analysis


The luminosities are 36.26, 41.37, 61.6 fb$^{-1}$ for the three runs. 

```python
#!/usr/bin/env python

import ROOT
import itertools

import os
import glob
infile = ROOT.TFile('../hists/BkgEstimate_crab__QCD_Pt-15to7000_TuneCP5_Flat_13TeV_pythia8_RunIIFall17NanoAODv4-PU2017_12Apr2018_Nano14Dec2018_102X.root')

cats = range(1,6)
vars = ['ttbarres/predJetMTTBARMod']
canvs = []
hists = []

for ivar,var in enumerate(vars):
    for icat in cats:
        cat = str(icat)
        iname = str(ivar)+'_'+str(icat)
        c = ROOT.TCanvas("c"+iname,"c"+iname )
        canvs.append(c)

        obs = infile.Get(var+cat)
        pred = infile.Get(var+cat+"_pred")
        pred.SetFillColor(ROOT.kYellow)
        obs.Draw()
        pred.Draw("same")
        obs.Draw("same")
        hists.append( [obs,pred])
        c.Draw()


```
